```python
def add(n1, n2):
    return n1 + n2

print(add(4, 5))
find_sum = add # In python function can also be stored in a variable
print(find_sum(4, 5))
```

```python
# functions are first class citizens
# function itself can be passed as an argument to another fn
# function itself can be returned from another fn

def add(n1, n2):
    return n1 + n2

def sub(n1, n2):
    return n1 - n2

def mul(n1, n2):
    return n1 * n2

def operate(n1, n2, operation): # operation(fn) -> callback fn, operate -> higher order function(hof)
    value = operation(n1, n2)
    return value

print(operate(10, 10, add))
print(operate(10, 10, sub))
print(operate(10, 10, mul))
```

```python
nums = [1, 2, 3, 4, 5, 6]

def get_factorial(num):
    fact = 1
    for i in range(2, num+1):
        fact = fact * i
    return fact

# fact_nums = []
# for num in nums:
#     factorial = get_factorial(num)
#     fact_nums.append(factorial)

# map(fn, list) -> new list[]
# fact_nums = map(get_factorial, nums)
# print(list(fact_nums))

# fact_nums = list(map(get_factorial, nums))
# print(fact_nums)

def get_even(num):
    return num % 2 == 0

# filter
even_nums = filter(get_even, nums)
print(list(even_nums))
```
