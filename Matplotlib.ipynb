```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# print(matplotlib.__version__)
```

```python
# basic plot
x = [3, 2, 1]
y = [1, 2, 3]

plt.plot(x, y)
plt.show()
```

```python
x = [2, 4, 6, 8, 10]
y = [1, 2, 3, 4, 5]

# resizing the figure
# plt.figure(figsize=(10, 5))
plt.figure(figsize=(5, 5), dpi=100) # dpi = dots per inch

plt.plot(x, y, color='#15ebb2', label='2*n', linestyle='-', linewidth=3, marker='*', markersize=10, markeredgecolor='black')
# short cut
# '[color][marker][line]'
# plt.plot(x, y, 'r.-')

plt.title('Doubling the data', fontdict={'fontname': 'Times New Roman', 'fontsize': 20})
plt.xlabel('experience', fontdict ={ 'fontname': 'consolas', 'fontsize': 15})
plt.ylabel('salary in lakhs', fontdict ={ 'fontname': 'consolas', 'fontsize': 15})

# plt.xticks([2, 4, 10, 20, 100, 38, 98])
# plt.yticks([1, 2, 3, 4, 5])

# plt.xticks(range(2, 11, 2))
# plt.yticks(range(1, 6))

# line 2
x2 = np.array([1, 2, 3, 4, 5])
plt.plot(x2, x2**3, label='n^3')

# save
plt.savefig('doubling.jpeg', dpi=500)

plt.legend()

plt.show()
```

```python
# bar chart
x = ['a', 'b', 'c']
y = [5, 2, 3]

plt.figure(figsize=(5, 5))
bar = plt.bar(x, y)
bar[0].set_hatch('/')
bar[1].set_hatch('*')
bar[2].set_hatch('.')
plt.title('bar chart')
plt.xlabel('character')
plt.ylabel('count')

plt.show()
```

```python
df = pd.read_csv('./mtcars.csv')
# df = df.dropna()
# df.shape
```

```python
x = range(32)
y = df['hp']

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='hp')
plt.title('hp plot')
plt.xlabel('index')
plt.ylabel('hp')

plt.legend()

plt.show()
```

```python
x = range(32)
y = df['disp']

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='disp')
plt.title('disp plot')
plt.xlabel('index')
plt.ylabel('disp')

plt.legend()

plt.show()
```

```python
x = range(32)
y1 = df['disp']
y2 = df['hp']

plt.figure(figsize=(6, 4))
plt.plot(x, y1, label='disp')
plt.plot(x, y2, label='hp')
plt.title('hp and disp plot')
plt.xlabel('index')
plt.ylabel('hp and disp')

plt.legend()

plt.show()
```

```python
x = df['hp']
y = df['disp']

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='hp vs disp')
plt.title('hp vs disp plot')
plt.xlabel('hp')
plt.ylabel('disp')

plt.legend()

plt.show()
```

```python
x = df['hp']
y = df['disp']

plt.figure(figsize=(6, 4))
plt.bar(x, y, label='hp vs disp')
plt.title('hp vs disp plot')
plt.xlabel('hp')
plt.ylabel('disp')

plt.legend()

plt.show()
```

```python
x = df['cyl']
y = df['wt']

plt.figure(figsize=(6, 4))
plt.bar(x, y, label='cyl vs wt')
plt.title('cyl vs wt plot')
plt.xlabel('cyl')
plt.ylabel('wt')

plt.legend()

plt.show()
```
